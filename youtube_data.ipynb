```python
# We have 2 index columns, so we will delete one:

df = df.drop(['index'], axis=1)

# Check if there is any duplicated rows:

duplicateRows = df[df.duplicated()]

duplicateRows

```

```python
# We found 19186 duplicated rows, lets make another data frame without them:

df1=df.drop_duplicates()
```

```python
# Set the date column ad a DATE type:

df1['videoPublished'] = pd.to_datetime(df1['videoPublished'])
df1.dtypes
```

```python
df1
```

```python
# The category column represented as numbers, I took the api of the explanation of any category, lets make another column with the category names:

# Create a dictionary mapping the category IDs to the names
category_names = {int(entry[0]): entry[1] for entry in [['2', 'Autos & Vehicles'], ['1', 'Film & Animation'],
             ['10', 'Music'], ['15', 'Pets & Animals'],
             ['17', 'Sports'], ['18', 'Short Movies'],
             ['19', 'Travel & Events'], ['20', 'Gaming'],
             ['21', 'Videoblogging'], ['22', 'People & Blogs'],
             ['23', 'Comedy'], ['24', 'Entertainment'],
             ['25', 'News & Politics'], ['26', 'Howto & Style'],
             ['27', 'Education'], ['28', 'Science & Technology'],
             ['29', 'Nonprofits & Activism'], ['30', 'Movies'],
             ['31', 'Anime/Animation'], ['32', 'Action/Adventure'],
             ['33', 'Classics'], ['34', 'Comedy'], ['35', 'Documentary'],
             ['36', 'Drama'], ['37', 'Family'], ['38', 'Foreign'],
             ['39', 'Horror'], ['40', 'Sci-Fi/Fantasy'], ['41', 'Thriller'],
             ['42', 'Shorts'], ['43', 'Shows'], ['44', 'Trailers']]}

# Use the "map" function to apply the dictionary to the videoCategoryId column
df1['category_name'] = df1['videoCategoryId'].map(category_names)
    
     
```

```python
# We will check whether the leading category (in terms of the number of videos) is the category that brings the creator the highest amount of interactions.

# Group the data by category and count the number of videos in each category:
category_counts = df1.groupby('category_name')['videoId'].nunique()

category_counts.plot.bar(stacked=True,figsize=(9,7))

# Add axis labels and a title
plt.xlabel('category_name')
plt.ylabel('Number of videos')
plt.title('number of videos in each category')
# Show the plot
plt.show()

#  Group the data by category and sum the views, likes, and dislikes for each category
category_sums = df1.groupby('category_name')['videoViewCount', 'videoLikeCount', 'VideoCommentCount'].sum()

# Divide the sums by the counts to get the averages
category_avgs = category_sums.divide(category_counts, axis=0)

# Plot the data as a stacked bar chart
category_avgs.plot.bar(stacked=True, figsize=(9,7))

# Add axis labels and a title
plt.xlabel('category_name')
plt.ylabel('Average Interactions')
plt.title('Average Interactions per Video by Category')
# Show the plot
plt.show()




```

```python
# explain- 

# As we showed previously, the category that is published the most is number 10 (music),
# but it is far from being the category that collects the most interactions, which is the "performance" category.

# We can think of a few reasons for it:
# 1. The "music" category may have a larger number of active channels that regularly upload new content. 
# 2. The "shows" category may include a smaller number of highly popular shows that drive a disproportionate amount of interactions. This could result in a higher average number of interactions per video in the "shows" category, even if there are fewer videos overall.
```

```python
# DF with all the channels without diplicates:
unique_channels1=df1.drop_duplicates("channelId")
```

```python
# Relationship between the total number of views for a channel and the number of subscribers:

# We will check if there is a connection between the popularity of the videos on the channel
# (total number of views on the channel) and the number of subscribers to the channel.

# Plot the number of views versus the number of subscribers
plt.scatter(unique_channels1['subscriberCount'], unique_channels1['channelViewCount'])

# Add axis labels and a title
plt.xlabel('Number of Subscribers')
plt.ylabel('Number of Views')
plt.title('Number of Views by Number of Subscribers')

# Show the plot
plt.show()

```

```python
# Lets also cheak the coralation-
corr = unique_channels1['channelViewCount'].corr(unique_channels1['subscriberCount'])
print(corr)

# explain- 

# I chose a scatter plot in order to show how much one variable is affected by another

# According to the resulting graph, it is difficult to determine exactly how much the two variables influence each other, 
# but it can be seen that there is indeed a strong relationship between them.

# That's why we will try to check also with the help of correlation.
# The correlation is: 0.7888077028423459
# which is a strong correlation, so we can certainly see that there is a connection between these two variables. 
```

```python
# We will check whether as the years go by the videos become shorter:

# We will check whether the videos become shorter as the years go by.
# (which makes sense since people prefer to receive information in as short a time as possible).

# Add Year column:
df1['year'] = df1['videoPublished'].dt.strftime('%Y')

# group the total videos mean time by years:
mean_elapsed_time_by_year=df1.groupby('year')['elapsedtime'].mean()

# Plot the Videos length over the years:
mean_elapsed_time_by_year.plot(figsize=(9,7))

# Add axis labels and a title:
plt.xlabel('Year')
plt.ylabel('video_length')
plt.title('Videos length over the years')

# Show the plot:
plt.show()
```

```python
# explain-
# We checked the average length of the videos in all the categories by year, we can see that there is indeed a constant decrease in the length of the videos every year,
# in a remarkably gradual way.
```

```python
#Now We will check now whether as the videos became shorter, so did the number of views on each video increases: 
unique_videos=df1.drop_duplicates("videoId")

# Sort the data by elapsed time
df_sorted = unique_videos.sort_values('elapsedtime', ascending=False)

# Plot the number of views versus elapsed time
plt.scatter(df_sorted['elapsedtime'], df_sorted['videoViewCount'])

# Add axis labels and a title
plt.xlabel('Elapsed Time')
plt.ylabel('Number of Views')
plt.title('Number of Views by Elapsed Time')

# Show the plot
plt.show()

#כבר לפי מראית עין בהסתכלות על הגרף ניתן לראות שאורך הסרטון לא בהכרח משפיע על מספר הצפיות שלו
```

```python
# explain- 

# We can see from the graph that the correlation between these 2 variables is not particularly strong.
# there is an increase in the number of views as the videos became shorter, but on a tiny scale.
```

```python
# Let's see the Correlation Matrix and spot few analysis:

sns.set(style="white")

corr = df1.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(7, 7))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.9, vmin=-0.9, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title("Correlation Matrix", fontsize=18)
```

```python
# explain- 

# Very strong positive correlation:

# channelViewCount - totalviews/channelElapsedTime
# views/elapsed_time - videoViewCount


# Strong positive linear correlation:

# video_dislike_count - video_view_count
# video_comment_count - video_like_count
# views/elapsed_time - video_dislike_count
# (And few more)
```
